```python
import pandas as pd

input_filename = 'input.xlsx'
df_input = pd.read_excel('./Rekap.xlsx', engine='openpyxl', sheet_name="Epoch 80")

class_names = ['bag', 'belt', 'boots', 'footwear', 'outer', 'dress', 'sunglasses', 'pants', 'top', 'shorts', 'skirt', 'headwear', 'scarf', 'undefined']
table_data = {class_name: [0] * (len(class_names)) for class_name in class_names}

df_table = pd.DataFrame(table_data, index=class_names)


for i in range(len(df_input['ImageId'])):
    image_id = df_input['ImageId'][i]
    if pd.notnull(image_id):
        mask_rcnn_labels = set(str(df_input['GroundTruthLabel'][i]).split(', '))
        superpixel_labels = set(str(df_input['PredictedLabelUsingMaskR-CNN'][i]).split(', '))

        # mask_rcnn_labels = set(str(df_input['GroundTruthLabel'][i]).split(', '))
        # superpixel_labels = set(str(df_input['PredictedLabelUsingSuperpixel'][i]).split(', '))

        # mask_rcnn_labels = set(str(df_input['PredictedLabelUsingMaskR-CNN'][i]).split(', '))
        # superpixel_labels = set(str(df_input['PredictedLabelUsingSuperpixel'][i]).split(', '))

        if all(label in superpixel_labels for label in mask_rcnn_labels):
            for label in mask_rcnn_labels:
                df_table.loc[label, label] += 1
        else:
            for label in mask_rcnn_labels:
                if(label not in superpixel_labels):
                    if(label == "outer" or label == "dress" or label == "top"):
                        print('y')
                        if("dress" in superpixel_labels):
                            df_table.loc[label, 'dress'] += 1
                            continue
                        elif("top" in superpixel_labels):
                            df_table.loc[label, 'top'] += 1
                            continue
                        elif("outer" in superpixel_labels):
                            df_table.loc[label, 'outer'] += 1
                            continue
                        elif("pants" in superpixel_labels):
                            df_table.loc[label, 'pants'] += 1
                            continue
                        else:
                            df_table.loc[label, 'undefined'] += 1
                    elif(label == "dress"):
                        print('y')
                        if("pants" in superpixel_labels):
                            df_table.loc[label, 'pants'] += 1
                            continue
                    elif(label == "footwear" or label == 'boots'):
                        print('y')
                        if("boots" in superpixel_labels):
                            df_table.loc[label, 'boots'] += 1
                            continue
                        elif("footwear" in superpixel_labels):
                            df_table.loc[label, 'footwear'] += 1
                            continue
                        else:
                            df_table.loc[label, 'undefined'] += 1
                    elif(label == "pants" or label == 'shorts' or label == 'skirt'):
                        print('y')
                        if("pants" in superpixel_labels):
                            df_table.loc[label, 'pants'] += 1
                            continue
                        elif("shorts" in superpixel_labels):
                            df_table.loc[label, 'shorts'] += 1
                            continue
                        elif("skirt" in superpixel_labels):
                            df_table.loc[label, 'skirt'] += 1
                            continue
                        else:
                            df_table.loc[label, 'undefined'] += 1
                    else:
                        df_table.loc[label, 'undefined'] += 1
                else:
                    df_table.loc[label, 'undefined'] += 1

output_filename = 'output.xlsx'
with pd.ExcelWriter(output_filename) as writer:
    df_table.to_excel(writer, sheet_name='Sheet1')

print("selesai")
```
